```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("C:\\Local Disk (A)\\Trent University\\Data Visualization\\Project\\Final Project\\HR_Analytics.csv")

# Convert necessary attributes
data['AgeGroup'] = data['AgeGroup'].astype('category')
data['Attrition'] = data['Attrition'].astype('category')

# Trellis Plot
sns.set(style="whitegrid")
g = sns.catplot(x="AgeGroup", y="JobSatisfaction", hue="Attrition", col="Attrition", data=data, kind="box", height=4, aspect=1)
g.set_axis_labels("Age Group", "Job Satisfaction")
g.set_titles("Job satisfaction by Age Group and Attrition")
g.set_xticklabels(rotation=0, ha="right")
plt.show()
```

```python
# In Attrition "No" group, Most employees have a Job Satisfaction level between 3 and 4, regardless of their age group.
# In Attrition "Yes" group, employees between 26-35 and 36-45 have the satisfaction within 3 or less.
# This indicates that, Higher Job Satisfaction levels are generally associated with employees who did not leave the company. 
# However, There may be a potential relationship between lower Job Satisfaction and higher attrition rates, particularly among younger employees.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("C:\\Local Disk (A)\\Trent University\\Data Visualization\\Project\\Final Project\\HR_Analytics.csv")

# Convert necessary attributes
data['Attrition'] = data['Attrition'].astype('category')
data['YearsSinceLastPromotion'] = data['YearsSinceLastPromotion'].astype('category')

# Stacked bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='YearsSinceLastPromotion', hue='Attrition', palette='viridis')
plt.title('Years Since Last Promotion by Attrition')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
# A significant number of employees who left the company (Attrition = Yes) have 0 years since their last promotion. 
# This indicates that many employees left shortly after being promoted.
# Employees who have not been promoted for a long time (e.g., more than 4 years) generally show fewer cases of attrition. 
# This suggests that long-term employees might be more stable and less likely to leave, even without recent promotions.
# There are small peaks in attrition at 1 and 2 years since the last promotion. 
# This could indicate critical periods where employees reassess their career progression and decide to leave if they do not see further advancement
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("C:\\Local Disk (A)\\Trent University\\Data Visualization\\Project\\Final Project\\HR_Analytics.csv")

# modifying "PerformanceRating" attribute
data['PerformanceCategory'] = data['PerformanceRating'].apply(lambda x: 'High Performance' if x > 3 else 'Low Performance')

# Box plot 
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='PerformanceCategory', y='PercentSalaryHike')
plt.title('Performance Category vs Percent Salary Hike')
plt.xlabel('Performance Category')
plt.ylabel('Percent Salary Hike')
plt.show()
```

```python
# The median percent salary hike for low performers is around 14%.
# The median percent salary hike for high performers is significantly higher, around 22%.
# This suggest that employees with higher perfromance recieve more salay compared to lower performing employees.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("C:\\Local Disk (A)\\Trent University\\Data Visualization\\Project\\Final Project\\HR_Analytics.csv")

# Bar Plot
sns.catplot(data=data, x='OverTime', hue='Attrition', kind='count', height=6, aspect=1.5)
plt.title('Bar Plot of Attrition vs OverTime')
plt.xlabel('OverTime')
plt.ylabel('Employee Count')
plt.xticks(ticks=[0, 1], labels=['No Overtime', 'Overtime'])
plt.show()
```

```python
# this indicates that overtime is correlated to attrition. 
# Most of the Employees who did not do overtime work, stayed at the company.
# This could also mean that overtime can cause employeees to leave the company.
```
